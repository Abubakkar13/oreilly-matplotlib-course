In this lesson we will use the basemap library to 

```python
%matplotlib inline

import json

import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
with open('./data/sfpd.json') as fin:
    data = json.load(fin)['data'];
    
lon = []
lat = []
for row in data:
    lat.append(float(row[-2][1]))
    lon.append(float(row[-2][2]))
```

```python
map = Basemap(urcrnrlat=37.821981,     # top
              urcrnrlon=-122.343189,   # bottom
              llcrnrlat=37.697940,     # left
              llcrnrlon=-122.574915)   # right
x, y = map(lon, lat)
```

```python
dpi = 96
width = 20
height = 10
plt.figure(figsize=(width, height))
map.arcgisimage(service='World_Topo_Map', xpixels=dpi*width)
map.scatter(x, y, s=15, c='r', marker='s', alpha=0.01, linewidths=0);
plt.show();
```
